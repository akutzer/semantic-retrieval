```python
import os
import json
os.makedirs(human_qa_path + 'final/', exist_ok=True)
witcher = []
harry = []
elder = []
j = 0
for file in os.listdir(human_qa_path):
    if file.endswith(".json") and 'blue' in file:  
        print(file)
        i = 0
        # Read the input JSON file
        try:
            with open(human_qa_path + file, 'r') as file2:
                data = json.load(file2)

        # Convert the data to the desired format
        
            for key, value in data.items():
                if value["pos_query"] == ['...', '...', '...']:
                    continue
                
                if value["pos_query"] == [[]]:
                    continue

                if value["pos_query"] == []:
                    continue
            
                j = j + 1
                if len(value["pos_query"]) == 3:

                    item = {
                        "id": j,
                        "revid": 0,  # You can assign a value for "revid" if applicable
                        "url": value["url"],
                        "title": value["title"],
                        "text": [value["passage"]],
                        "positive": [value["pos_query"]],
                        "negative": [value["neg_query"]]
                    }
                else:
                    item = {
                        "id": int(key),
                        "revid": 0,  # You can assign a value for "revid" if applicable
                        "url": value["url"],
                        "title": value["title"],
                        "text": [value["passage"]],
                        "positive": value["pos_query"],
                        "negative": value["neg_query"]
                    }
                if i > 34:
                    continue 
                i = i + 1 
                        
                if 'harry' in file:
                    harry.append(item)
                elif 'elder' in file:
                    elder.append(item)
                elif 'witcher' in file:
                    witcher.append(item)
            print(i)
        except Exception as e:
            print('error in file ' + file)
            print(e)

        
with open(human_qa_path + 'final/' + 'witcher.json', 'w') as file:
    json.dump(witcher, file, indent=4)

with open(human_qa_path + 'final/' + 'elder.json', 'w') as file:
    json.dump(elder, file, indent=4)

with open(human_qa_path + 'final/' + 'harry.json', 'w') as file:
    json.dump(harry, file, indent=4)
```

```python
fandoms_qa_path2 = '../../data/humanfandoms/final/fandoms_all/all/'
df_passages_all = pd.read_csv(fandoms_qa_path1 + 'passages.tsv', sep ='\t')
df_passages_all
df_passages_human = pd.read_csv(fandoms_qa_path2 + 'passages.tsv', sep ='\t')
df_passages_human

df_triples_human = pd.read_csv(fandoms_qa_path2 + 'triples.tsv', sep ='\t')


```

```python
map_pid = {}
map_wid = {}

from tqdm import tqdm

j = 0
for ind, row in tqdm(df_passages_human.iterrows(), total=df_passages_human.shape[0]):
    # j = j + 1
    # if j >= 10:
    #     pid = -1
    #     wid = -1
    #     map_pid[row['PID']] = pid
    #     map_wid[row['WID']] = wid
    #     continue
    try:
        selected = (df_passages_all[ df_passages_all['passage'].str.contains(row['passage'][int(len(row['passage']) / 2) - 20:int(len(row['passage']) / 2) + 20 ])] )
        if  len(selected['PID'].values) == 0:
            pid = -1
            wid = -1
            
        else:
            pid = selected['PID'].values[0]
            wid = selected['WID'].values[0]
    except:
        pid = -1
        wid = -1
        print('error')

        
    map_pid[row['PID']] = pid
    map_wid[row['WID']] = wid
    
```

```python
df_triples_human['PID'] = df_triples_human['PID'].apply(lambda row: map_pid[row])
df_triples_human
```
